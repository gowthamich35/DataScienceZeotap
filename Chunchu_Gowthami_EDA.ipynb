*Insight-1*

**The total number of unique customers is calculated by using the nunique() method on the CustomerID column of the customers.csv file. This method counts all distinct customer IDs, which gives us the overall customer base.**

```python
#insight-1
#Total Number of Customers
total_customers = cs['CustomerID'].nunique()
print(f"Total Number of Customers: {total_customers}")
```

*Insight-2*

**To find the number of unique products sold, we use the nunique() method on the ProductID column from the transactions.csv file. This identifies how many different products are present in the transactions.**

```python
#insight-2
#Number of Products Sold
unique_products_sold = ts['ProductID'].nunique()
print(f"Number of Products Sold: {unique_products_sold}")
```

*Insight-3*

**By using the idxmax() method, we identify the product with the highest price from the products.csv file. The method returns the index of the row with the maximum price, which we use to retrieve the product details.**

```python
# 3. Most Expensive Product
most_expensive_product = ps.loc[ps['Price'].idxmax()]
print(f"Most Expensive Product: {most_expensive_product['ProductName']} with a price of {most_expensive_product['Price']} USD")
```

*Insight-4*

**Using the groupby('CustomerID') method on the transactions.csv file, we group the data by customer, and then use size() to count the number of transactions (orders) for each customer.**

```python
# Number of Orders per Customer
orders_per_customer = ts.groupby('CustomerID').size()
print(f"Number of Orders per Customer:")
print(orders_per_customer)
```

*Insight-5*

**To find the customer who placed the most orders, we identify the CustomerID associated with the maximum value in the orders_per_customer series. We use orders_per_customer.max() to get the maximum order count, then filter to get the corresponding customer.**

```python
#CustomerID with the maximum number of orders
max_orders_customer = orders_per_customer[orders_per_customer == orders_per_customer.max()]
print(f"Customer(s) with the Maximum Orders: {max_orders_customer}")
```
